```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("loan.csv", low_memory=False)
pd.set_option('display.max_rows', None)
pd.set_option("display.max_columns", None)
df.head().T
```

```python
df.shape
```

```python
df.describe()
```

```python
#check for total no of null values in each column 
100*(df.isnull().sum()/df.shape[0])
```

```python
missing_threshold = 0.7 * len(df)
df_cleaned = df.dropna(thresh=missing_threshold, axis=1)
not_reqd = ["member_id","policy_code","url","zip_code","emp_title","title","last_pymnt_d",
            "out_prncp","out_prncp_inv","revol_bal","pymnt_plan","issue_d","initial_list_status"]
df_cleaned = df_cleaned.drop(not_reqd,axis=1)
```

```python
df_cleaned.shape
```

```python
df_cleaned.info()
```

```python
# The applicants with loan_status = current indicate already approved applications. So, can be ignored for the current analysis
# Drop all the records with loan_status = current
df_cleaned = df_cleaned[df_cleaned['loan_status'] != 'Current']
```

```python
100*(df_cleaned.isnull().sum()/df_cleaned.shape[0])
```

```python
# filling the null values with 0 assuming that the borrower hasn't worked many years for his data to be recorded. Also, using regex to extract the number of years from all of the data.
# replacing percentage sign in revol_util column
```

```python
df_cleaned['emp_length'].fillna(value=0,inplace=True)
df_cleaned['emp_length'].replace(to_replace='[^0-9]+', value='', inplace=True, regex=True)
df_cleaned['emp_length'] = pd.to_numeric(df_cleaned['emp_length'], errors='coerce')
df_cleaned["revol_util"].replace(r'%', '', regex=True, inplace=True)
df_cleaned['revol_util'] = pd.to_numeric(df_cleaned['revol_util'], errors='coerce')
df_cleaned["int_rate"].replace(r'%', '', regex=True, inplace= True)
df_cleaned['int_rate'] = df_cleaned['int_rate'].replace('N/A', pd.NA)
df_cleaned['int_rate'] = df_cleaned['int_rate'].astype(float)
```

### Univariate Analysis

```python
# Plot the distribution of loan amount
sns.histplot(df_cleaned["loan_amnt"], bins=10)
plt.title("Distribution of Loan Amount")
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of interest rates
sns.displot(df_cleaned["int_rate"], bins=20, kde=True)
plt.title("Distribution of Interest Rates")
plt.xlabel('Interest Rates')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the distribution of Loan Status
sns.countplot(data=df_cleaned, x='loan_status', order=df_cleaned['loan_status'].value_counts().index)
plt.title('Distribution of Loan Statuses')
plt.xlabel('Loan Status')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the distribution of Employment length in years
current_labels = df_cleaned['emp_length'].value_counts().index
sns.countplot(data=df_cleaned, x='emp_length', order=current_labels)
plt.title('Distribution of Employment length in years.')
plt.xlabel('Employment Length')
plt.ylabel('Frequency')
formatted_labels = [f'{label} year' if label == 1 else f'{label} years' for label in current_labels]
plt.xticks(ticks=range(len(current_labels)), labels=formatted_labels, rotation=45)
plt.show()
```

```python
# Plot the distribution of Home Ownership
current_labels = df_cleaned['home_ownership'].value_counts().index
sns.countplot(data=df_cleaned, x='home_ownership', order=current_labels)
plt.title('Distribution of Home Ownership')
plt.xlabel('Home Ownership')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the distribution of Loan Purpose
loan_purpose_counts = df_cleaned['purpose'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=loan_purpose_counts.values, y=loan_purpose_counts.index, palette='Set2')
plt.title('Distribution of Loan Purposes')
plt.xlabel('Number of Loans')
plt.ylabel('Loan Purpose')
plt.show()
```

```python
# Plot the distribution of Loan Terms
plt.figure(figsize=(5, 5))  # Set figure size
plt.title('Distribution of Loan Terms', fontsize=16, fontweight='bold', color='#2c3e50')  # Title

# Plot the pie chart with custom styling
df_cleaned.term.value_counts().plot.pie(
    autopct='%1.1f%%',      # Display percentage on each slice
    colors=['#3498db', '#e74c3c'],  # Custom colors
    startangle=90,          # Start from the top
    shadow=True,            # Add a shadow effect
    explode=(0.05, 0),      # Slightly separate the first slice
    labels=['36 months', '60 months'],  # Custom labels
    wedgeprops={'edgecolor': 'black'}  # Add black edge to slices
)
plt.ylabel('')  # Remove default y-label
plt.show()
```

```python
# Plot the Distribution of Monthly Installments
plt.figure(figsize=(10, 5))  # Set figure size
plt.title('Distribution of Monthly Installments', fontsize=16, fontweight='bold', color='#2c3e50')  # Custom title
plt.xlabel('Installment per Month (USD)', fontsize=12, fontweight='bold', color='#2c3e50')  # Custom X-axis label
plt.ylabel('Frequency', fontsize=12, fontweight='bold', color='#2c3e50')  # Custom Y-axis label

# Plot the histogram with custom styling
df_cleaned.installment.plot.hist(
    bins=30,                 # Number of bins
    color="#e74c3c",         # Bar color
    edgecolor="black",       # Edge color of bars
    alpha=0.7                # Set transparency
)

plt.grid(True, linestyle='--', alpha=0.6)  # Add a grid
plt.xticks(fontsize=10, color='#3498db')   # Customize x-tick labels
plt.yticks(fontsize=10, color='#3498db')   # Customize y-tick labels

plt.show()
```

```python
# Plot the Distribution of Customer DTI ratio
plt.figure(figsize=(10, 5))  # Set the figure size
plt.title('Distribution of Customer DTI (Debt-to-Income Ratio)', fontsize=16, fontweight='bold', color='#2c3e50')  # Custom title
plt.xlabel('DTI Ratio', fontsize=12, fontweight='bold', color='#2c3e50')  # Custom X-axis label
plt.ylabel('Frequency', fontsize=12, fontweight='bold', color='#2c3e50')  # Custom Y-axis label

# Plot the histogram with custom styling
df_cleaned.dti.plot.hist(
    bins=25,                 # Number of bins
    color="#e74c3c",         # Bar color
    edgecolor="black",       # Edge color of bars
    alpha=0.7                # Transparency level for the bars
)

plt.grid(True, linestyle='--', alpha=0.5)  # Add a grid with a light dashed line
plt.xticks(fontsize=10, color='#34495e')   # Customize X-axis ticks
plt.yticks(fontsize=10, color='#34495e')   # Customize Y-axis ticks
plt.xlim(left=0)  # Start x-axis from 0 for better visualization

plt.show()
```

```python
# Plot the Verification Status of Loan
plt.title('Verification Status of Loan', fontsize=14)  # Simple title
plt.xlabel('Count', fontsize=12)  # X-axis label

# Plot the horizontal bar chart
df_cleaned.verification_status.value_counts().plot.barh(
    color="#e74c3c"  # Bar color
)

plt.grid(axis='x', linestyle='--', alpha=0.6)  # Add a simple grid only on x-axis

plt.show()
```

```python
# Plot the Loan Grade Analysis
plt.title('Loan Grade Analysis', fontsize=14)  # Simple title
plt.xlabel('Count', fontsize=12)  # X-axis label

# Plot the horizontal bar chart with some customization
df_cleaned.grade.value_counts().plot.barh(
    color="#e74c3c",  # Custom bar color
    edgecolor="black"  # Add edge color to the bars
)

plt.grid(axis='x', linestyle='--', alpha=0.5)  # Add gridlines to the x-axis for better readability
plt.xticks(fontsize=10)  # Customize x-ticks
plt.yticks(fontsize=10)  # Customize y-ticks

plt.tight_layout()  # Adjust layout to ensure everything fits well
plt.show()
```

```python
# Plot the Sub-Grade Distribution
plt.title('Sub-Grade Distribution', fontsize=16)  # Title with slightly larger font
plt.xlabel('Sub-Grade', fontsize=12)  # X-axis label
plt.ylabel('Count', fontsize=12)  # Y-axis label

# Plot the bar chart with basic customization
df_cleaned.sub_grade.value_counts().plot.bar(
    color="#e74c3c",  # Bar color
    edgecolor="black"  # Add black edges to bars
)

plt.xticks(rotation=45, fontsize=10)  # Rotate x-ticks slightly for readability
plt.yticks(fontsize=10)  # Customize y-ticks

plt.tight_layout()  # Ensure everything fits well
plt.show()
```
